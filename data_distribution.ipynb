```python
# data distribution

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# loas the data from seaborn
df = sns.load_dataset('tips')
df.head()
```

```python
# draw a histplot curve with kde
sns.histplot(data=df, x = 'total_bill',kde=True)
plt.figure(figsize=(8,6))
```

```python
sns.histplot(data=df, x = 'size',kde=True)
plt.figure(figsize=(8,6))
```

# Kurtosis and Skewness

```python
# find kurtosis of age 
df_third = df_01[df_01['class']=='third']
stats.kurtosis(df_third['fare'],fisher=False)
```

```python
stats.skew(df_01['fare'])
```
